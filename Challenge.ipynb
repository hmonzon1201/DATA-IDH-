```python
import random
import itertools
import pandas as pd

class Player:
    def __init__(self, name, score=0):
        self.name = name
        self.score = score

    def increase_score(self, value):
        self.score += value

class PlayerBetter(Player):
    def __init__(self, name, score=0):
        super().__init__(name, score)
        self.strategy = None

    def get_strategy(self):
        strategies = ['Rock', 'Paper', 'Scissors']
        self.strategy = random.choice(strategies)
        return self.strategy
```

```python
# Definir las estrategias y las reglas del juego
strategies = ['Rock', 'Paper', 'Scissors']
payoff = {
    ('Rock', 'Paper'): (0, 1),
    ('Paper', 'Rock'): (1, 0),
    ('Rock', 'Scissors'): (1, 0),
    ('Scissors', 'Rock'): (0, 1),
    ('Paper', 'Scissors'): (0, 1),
    ('Scissors', 'Paper'): (1, 0),
    ('Rock', 'Rock'): (0, 0),
    ('Paper', 'Paper'): (0, 0),
    ('Scissors', 'Scissors'): (0, 0)
}

# Nombres de los jugadores
names = ['Jim', 'Jane', 'Peter', 'Zoe', 'Edgar', 'Jose', 'Charles', 'Pedrito', 'Ricardo', 'Paco']

# Crear los jugadores
society = [PlayerBetter(name) for name in names]
```

```python
# Resultados
round_results = []

# 100 rondas
for aRound in range(100):

    for player1, player2 in itertools.combinations(society, 2):
        player1.get_strategy()
        player2.get_strategy()

        # Determinar el resultado
        result = payoff[(player1.strategy, player2.strategy)]

        # Actualizar el score
        player1.increase_score(result[0])
        player2.increase_score(result[1])


    round_results.append(pd.DataFrame([{'name': p.name, 'score': p.score, 'strategy': p.strategy} for p in society]))


final_society = pd.DataFrame([{'name': p.name, 'score': p.score, 'strategy': p.strategy} for p in society])
print(final_society)

#Ganadores
winner_score = final_society.score.max()
winners = final_society[final_society.score == winner_score]
print("Ganador(es):")
print(winners)

# Resultados en vivo
for round_num, result in enumerate(round_results):
    print(f"Resultados de la Ronda {round_num + 1}:")
    print(result)
    print("\n")
```
